5. <b>About Diabetes</b>

   Diabetes is a chronic disease that occurs either when the pancreas does not produce enough insulin or when the body cannot effectively use the insulin it produces. Insulin is a hormone that regulates blood glucose. Hyperglycaemia, also called raised blood glucose or raised blood sugar, is a common effect of uncontrolled diabetes and over time leads to serious damage to many of the body's systems, especially the nerves and blood vessels.

   <b>Types of Diabetes</b>

   The main three types of Diabetes are:

   a. <b>Type 1 Diabetes (Chronic)</b>

   Type 1 diabetes (previously known as insulin-dependent, juvenile or childhood-onset) is characterized by deficient insulin production and requires daily administration of insulin.It is an autoimmune condition where the body's immune system attacks and destroys insulin-producing beta cells in the pancreas.

   Main issues:
    - Insulin Dependency: Individuals require lifelong insulin therapy to manage blood glucose levels.
    - Hyperglycemia: High blood sugar levels can cause damage to organs and tissues over time.
    - Hypoglycemia: Risk of low blood sugar levels due to insulin therapy, which can cause confusion, seizures, and loss of consciousness.

   b. <b>Type 2 Diabetes</b>

   Type 2 diabetes affects how your body uses sugar (glucose) for energy. It stops the body from using insulin properly, which can lead to high levels of blood sugar if not treated. Factors that contribute to developing type 2 diabetes include being overweight, not getting enough exercise, and genetics. It is a condition characterized by insulin resistance and/or insufficient insulin production.

   Main Issues:
    - Insulin Resistance: The body's cells do not respond properly to insulin, leading to elevated blood glucose levels.
    - Progressive Nature: Often starts with insulin resistance and can progress to require insulin therapy as beta cells become less effective.
    - Lifestyle Factors: Strongly associated with obesity, poor diet, and lack of physical activity.

   c. <b> Gestational diabetes</b>

   Gestational diabetes is hyperglycaemia with blood glucose values above normal but below those diagnostic of diabetes. Gestational diabetes occurs during pregnancy.

   Main Issues:
    - Pregnancy Complications: Can lead to high birth weight, premature delivery, and increased risk of cesarean delivery.
    - Future Risk: Increases the mother’s risk of developing Type 2 diabetes later in life.

   Also, in our dataset we will classify people with Prediabetes which is a critical stage that serves as a warning sign for the potential development of Type 2 diabetes.

  <b> Symptoms of Diabetes</b>

    - feeling very thirsty
    - needing to urinate more often than usual
    - blurred vision
    - feeling tired
    - losing weight unintentionally

  <b> Risk Factors of Diabetes </b>

    - Cardiovascular Disease: Increased risk of heart disease, stroke, and hypertension due to damage to blood vessels and nerves.
    - Neuropathy: Damage to nerves, particularly in the extremities, causing pain, tingling, and loss of sensation.
    - Nephropathy: Kidney damage leading to chronic kidney disease or kidney failure, necessitating dialysis or transplantation.
    - Retinopathy: Damage to the blood vessels in the retina, potentially causing blindness.
    - Foot Problems: Poor circulation and nerve damage can lead to foot ulcers and infections, sometimes requiring amputation.
    - Skin Conditions: Higher susceptibility to bacterial and fungal infections.
    - Mental Health: Increased risk of depression and anxiety due to the chronic nature of the disease and its complications.

  <b> Monitoring Diabetes </b>

     - Blood sugar monitoring: Monitoring your blood sugar (glucose) is key to determining how well your current treatment plan is working.
      - Oral diabetes medications: Oral diabetes medications (taken by mouth) help manage blood sugar levels in people who have diabetes but still produce some insulin — mainly people with Type 2 diabetes and prediabetes.
      - Insulin: People with Type 1 diabetes need to inject synthetic insulin to live and manage diabetes. Some people with Type 2 diabetes also require insulin
      - Diet: Meal planning and choosing a healthy diet for you are key aspects of diabetes management, as food greatly impacts blood sugar.
      - Exercise: Physical activity increases insulin sensitivity (and helps reduce insulin resistance), so regular exercise is an important part of management for all people with diabetes.

  <b> Prevention </b>
     There are several strategies which can be followed to control Diabetes:
   
    - Reach and keep a health body weight
    - Stay physically active with at least 30 minutes of moderate exercise each day
    - Eat a healthy diet and avoid sugar and saturated fat
    - Do Not smoke tobacco
    - Limit alcohol intake.
    - Get adequate sleep (typically 7 to 9 hours)
    - Take medications as directed by your healthcare provider

## Task 5 - Baselines Core Classifiers

A baseline model is a ground zero, uncomplicated model that may be used as a starting point for more intricate models

Source: https://medium.com/@preethi_prakash/understanding-baseline-models-in-machine-learning-3ed94f03d645

```python
# Evaluation Table Method taken from Class Notes
results = pd.DataFrame()
def add_eval(results, algorithm, y_train, y_train_pred, y_validate , y_validate_pred, y_test, y_test_pred):
    for dataset, actual, predicted in zip(("train", "validation","test"), (y_train,y_validate, y_test), (y_train_pred, y_validate_pred, y_test_pred)):
        mse=mean_squared_error(actual, predicted)
        results= pd.concat([results, pd.DataFrame([{
            "algorithm": algorithm,
            "dataset": dataset,
            "MSE": mse,
            "RMSE": math.sqrt(mse),
            "MAE": mean_absolute_error(actual, predicted),
            "Accuracy": accuracy_score(actual,predicted),
            "Recall": recall_score(actual,predicted, average = "macro") }])], ignore_index=True)
    return results
```

### First Architecture of Deep Learning Nerual Network : Multilayer Perceptron/ Feedforward Nerual Network

#### Defining of Classifier

```python
class MLPClassifier(nn.Module):
    def __init__(self, input_size, output_size, num_neurons):
        super(MLPClassifier, self).__init__()
        self.hidden_1 = nn.Linear(input_size, num_neurons)
        self.activation = nn.ReLU()
        self.hidden_2 = nn.Linear(num_neurons,num_neurons)
        self.hidden_3 = nn.Linear(num_neurons,num_neurons)
        self.output = nn.Linear(num_neurons, output_size)

    def forward(self, x):
        out1 = self.hidden_1(x)
        out2 = self.activation(out1)
        out3 = self.hidden_2(out2)
        out4 = self.activation(out3)
        out5 = self.hidden_3(out4)
        out6 = self.output(out5)
        return out6
```

#### Traning the model

```python
# Setting parameters
input_size = X_train_under_tensor.shape[1]
output_size = 3
num_neurons = 128
learning_rate = 0.01
num_epochs = 200

# Initialize the model, loss function, and optimizer
model_MLP = MLPClassifier(input_size= input_size, output_size= output_size, num_neurons= num_neurons)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model_MLP.parameters(), lr=learning_rate)

best_val_accuracy = 0
best_model_state = None

#Evaluation for epochs
for epoch in range(num_epochs):
        model_MLP.train()
        y_train_pred = model_MLP(X_train_under_tensor)

        train_loss = criterion(y_train_pred, y_train_under_tensor)
        optimizer.zero_grad()

        train_loss.backward()

        optimizer.step()

        # Validation
        model_MLP.eval()
        with torch.no_grad():
          val_outputs = model_MLP(X_val_tensor)
          _, val_predicted = torch.max(val_outputs.data, 1)
          val_accuracy = accuracy_score(y_val_tensor.numpy(), val_predicted.numpy())
      
          if val_accuracy > best_val_accuracy:
            best_val_accuracy = val_accuracy
            best_model_state = model_MLP.state_dict()
      
          if ((epoch+1) % 20 == 0):
             print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {train_loss.item():.4f}, Validation Accuracy: {val_accuracy:.4f}')

# Load the best model state
model_MLP.load_state_dict(best_model_state)       
```

### Second Architecture of Deep Learning Network: Convolutional Neural Network

```python
# Reshape the data to fit the CNN input requirements
X_train_under_unsquezed = X_train_under_tensor.unsqueeze(1)  # Add a channel dimension
X_test_unsqueezed = X_test_tensor.unsqueeze(1)    
X_val_unsqueezed = X_val_tensor.unsqueeze(1)


class CNNClassifier(nn.Module):
    def __init__(self, input_channels, num_classes, K=2):
        super(CNNClassifier, self).__init__()

        self.conv1 = nn.Conv1d(input_channels, 16, kernel_size=3, padding=1, bias=False)

        self.conv2 = nn.Conv1d(16, 32, kernel_size=3, padding=1, bias=False)

        self.pool = nn.MaxPool1d(kernel_size=2, stride=2)

        self.fc1 = nn.Linear(32 * (X_train_under_unsquezed.shape[2] // 4), 128)

        self.fc2 = nn.Linear(128, num_classes)

        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.conv1(x)
        x = self.relu(x)
        x = self.pool(x)
        x = self.conv2(x)
        x = self.relu(x)
        x = self.pool(x)
        x = x.view(x.size(0), -1) 
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
```

#### Traning the model

```python
# Set parameters
input_channels = 1
num_classes = 3
learning_rate = 0.01
num_epochs = 200

# Initialize the model, loss function, and optimizer
model_CNN = CNNClassifier(input_channels, num_classes)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model_CNN.parameters(), lr=learning_rate)

#Evaluation for epochs
for epoch in range(num_epochs):
        model_CNN.train()
        y_train_pred = model_CNN(X_train_under_unsquezed)

        train_loss = criterion(y_train_pred, y_train_under_tensor)
        optimizer.zero_grad()

        train_loss.backward()

        optimizer.step()

        if ((epoch+1) % 20 == 0):
              print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {train_loss.item():.4f}')
```

#### Model Evaluation

```python
model_CNN.eval()
with torch.no_grad():
    train_outputs = model_CNN(X_train_under_unsquezed)
    _, train_predicted = torch.max(train_outputs.data, 1)

    val_outputs = model_CNN(X_val_unsqueezed)
    _, val_predicted = torch.max(val_outputs.data, 1)

    test_outputs = model_CNN(X_test_unsqueezed)
    _, test_predicted = torch.max(test_outputs.data, 1)

    results = add_eval(results, "CNN Architecture",
                   y_train_under_tensor, train_predicted, y_val_tensor, val_predicted,
                   y_test_tensor, test_predicted)
results
```

```python
class MLPClassifierLayers(nn.Module):

 def __init__(self, input_dim, output_dim, num_layers, num_nuerons):
        super(MLPClassifierLayers, self).__init__()
        layers = []
        layers.append(nn.Linear(input_dim, num_nuerons))
        layers.append(nn.ReLU())

        for _ in range(num_layers - 1):
            layers.append(nn.Linear(num_nuerons, num_nuerons))
            layers.append(nn.ReLU())
        layers.append(nn.Linear(num_nuerons, output_dim))
        self.model = nn.Sequential(*layers)

 def forward(self, x):
        return self.model(x)
```

```python
# Setting parameters
input_size = X_train.shape[1]
output_size = 3
num_neurons = 128
learning_rate = 0.01
num_epochs = 200
num_layers = [3,4,5]

best_val_loss = float('inf')

for i in num_layers:

    model_MLP_layers = MLPClassifierLayers(input_dim= input_size, output_dim= output_size, num_layers = i, num_nuerons=num_neurons )
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model_MLP_layers.parameters(), lr=learning_rate)

    print("Training first for num of layers =", i)
    for epoch in range(num_epochs):
        model_MLP_layers.train()
        y_train_pred = model_MLP_layers(X_train_under_tensor)
        train_loss = criterion(y_train_pred, y_train_under_tensor)

        val_pred = model_MLP_layers(X_val_tensor)
        val_loss = criterion(val_pred, y_val_tensor)

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_hidden_layers = i

        optimizer.zero_grad()

        train_loss.backward()

        optimizer.step()

        if ((epoch+1) % 20 == 0):
          print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {train_loss.item():.4f}')

print("best Hidden Layer Number is :", best_hidden_layers)
```

```python
# increasing number of convultional layers to 3 from 2 in previous case
class CNNClassifierActivation(nn.Module):
    def __init__(self, input_channels, num_classes, activation_fun):
        super(CNNClassifierActivation, self).__init__()

        self.conv1 = nn.Conv1d(input_channels, 16, kernel_size=3, padding=1, bias=False)

        self.conv2 = nn.Conv1d(16, 32, kernel_size=3, padding=1, bias=False)

        self.pool = nn.MaxPool1d(kernel_size=2, stride=2)

        self.conv3 = nn.Conv1d(32, 64, kernel_size=3, padding=1, bias=False)

        self.fc1 = nn.Linear(64 , 128)

        self.fc2 = nn.Linear(128, num_classes)

        self.relu = activation_fun

    def forward(self, x):
        x = self.conv1(x)
        x = self.relu(x)
        x = self.pool(x)
        x = self.conv2(x)
        x = self.relu(x)
        x = self.pool(x)
        x = self.conv3(x)
        x = self.relu(x)
        x = self.pool(x)
       
        x = x.view(x.size(0), -1)  # Flatten the output for the fully connected layer
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
```

### Explanation of Employed Methodology for final Model

I have used neural network architecture feed forward method: 

A feed-forward neural network (FFNN) is a type of artificial neural network where the connections between the nodes do not form a cycle

Components of a Feed-Forward Neural Network:

1. Neurons : The basic units of the network. Each neuron receives input, processes it, and produces an output.
2. Layers : The first layer that receives the input data. Each neuron in this layer represents an input feature. Intermediate layers between the input and output layers called hidden layers.  The final layer that produces the network's output is the output layer.
3. Weights: Parameters that connect neurons in one layer to neurons in the next layer
4. Biases : Additional parameters added to the neurons' inputs to allow the activation function to be shifted to the left or right, which helps the network model more complex patterns
5. Activation Functions: Functions applied to the input of a neuron to produce the output

Working of a Feed-Forward Neural Network: 

1. Forward Propagation: The process where input data is passed through the network layer by layer, from the input layer, through the hidden layers, to the output layer.
2. Loss Function: A function that measures the difference between the network's output and the actual target values
3. Backpropagation: The process used to update the weights and biases in the network.
4. Training:The network is trained by iteratively performing forward propagation and backpropagation over many epochs (complete passes through the training dataset). During this process, the network adjusts its weights and biases to minimize the loss function.
